# DATA VISUALIZATION USING PHYTON

Data visualization using Python is the process of visualizing data using various libraries and tools available in the Python programming environment. Data visualization is a way to present information and patterns in data through graphical representations, such as charts, plots, maps, and more. The goal of data visualization is to clarify and communicate information in an easily understandable and engaging manner.

Some commonly used libraries for performing data visualization in Python include:

1. **Matplotlib**: It is the most popular and versatile library for data visualization in Python. Matplotlib provides various types of plots such as line plots, scatter plots, bar plots, histograms, and more.

2. **Seaborn**: Built on top of Matplotlib, Seaborn provides a higher-level interface for creating more aesthetically pleasing and complex plots. Seaborn is suitable for statistical data exploration and visualization.

3. **Plotly**: It is a powerful library for creating interactive plots and data visualization. Plotly supports various types of plots and interactivity that can be integrated with various platforms.

4. **Folium**: It is a library for creating interactive maps using geographic data. Folium allows you to add markers, polygons, and other interactions to maps.

5. **Pandas Visualization**: It is an integrated data visualization module within the Pandas library. Pandas has methods for creating plots directly from its DataFrame and Series objects.

6. **Altair**: It is a declarative library that allows for deep visualization with simple and intuitive syntax.

Through these libraries, Python users can create informative, engaging, and often interactive data visualizations to explore and understand data better. With data visualization, users can identify patterns, trends, anomalies, and other insights that may be hidden in the data.

## Using Seaborn

Seaborn is a library for creating graphs and statistics using Python. This library is built based on the existing Matplotlib library. Then integrated with the data structure in Pandas.

### Example

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds_df = sns.load_dataset ("diamonds")
diamonds_df.head()
```

```python
# Stacked Bar Chart

sns.barplot(x='cut',
            y='price',
            data=diamonds_df)
```

We can draw a bar chart to visualize how the average bill amount varies on different days of the week. We can do this by calculating the daily average and then using plt bars. seaborn library also provides a barplot function that can calculate the average automatically

```python
# Verical Pivot

plt.figure(figsize=(12, 8))
sns.barplot(x='cut', y='price', hue='color', data=diamonds_df)
```

```python
# Horizontal Pivot

plt.figure(figsize=(12, 8))
sns.barplot(x='price', y='cut', hue='color', data=diamonds_df)
```

## Histogram

A histogram is a graphical display of frequency tabulation depicted with bar graphics as a manifestation of data binning. Each bar display shows the proportion of frequencies in each series of adjacent categories with non-overlapping intervals.

### Example

```python
healthexp_df = sns.load_dataset("healthexp")
healthexp_df.head()                          
```

```python
# Make a histogram from the data above

plt.title("Distribution of Life Expectancy")
plt.hist(healthexp_df.Life_Expectancy);
```

```python
# Specifying the Number Of  Bins

plt.hist(healthexp_df.Life_Expectancy, bins=13);
```

```python
# Make a Range

import numpy as np

plt.hist(healthexp_df.Life_Expectancy, bins=np.arange(70, 87, 0.75));
```
