# BASIC DATA VISUALIZATION AND COMMUNICATION

Basic data visualization and communication in data analytics refers to the ability to depict and present information from data in graphical forms that are easy to understand. It involves using charts, graphs, tables, and other data visualization tools to communicate findings or patterns discovered in data in a clear and effective manner. The goal is to make complex information more understandable and digestible to diverse audiences.

## 1. Comparison

Comparison is the process of examining two or more things to determine their differences, similarities, or relationships. In the context of data analytics, comparison can be done between various metrics, variables, or data groups to gain a better understanding of existing patterns or trends.

```python
# Example Bar Chart Vertical

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

# Creating the bar chart
plt.bar(categories, values)

# Adding labels to the x and y axes as well as the title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart Example')

# Displaying the chart
plt.show()
```

In this example, we use the Matplotlib library to create a vertical bar chart. The category data is stored in the categories list, while the corresponding values are stored in the values list. The plt.bar() function is used to create the bar chart, and then the plt.xlabel(), plt.ylabel(), and plt.title() functions are used to add labels to the x-axis, y-axis, and title of the chart respectively. Finally, the plt.show() function is used to display the chart. You can adjust the data and visualization parameters as needed.

```python
# Example Bar Chart Horizontal

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

# Creating the horizontal bar chart
plt.barh(categories, values)

# Adding labels to the x and y axes as well as the title
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart Example')

# Displaying the chart
plt.show()
```

In this example, we use the Matplotlib library to create a horizontal bar chart. The category data is stored in the categories list, while the corresponding values are stored in the values list. The plt.barh() function is used to create the horizontal bar chart instead of plt.bar(). Then, we add labels to the x-axis, y-axis, and title of the chart using the plt.xlabel(), plt.ylabel(), and plt.title() functions respectively. Finally, the plt.show() function is used to display the chart. You can adjust the data and visualization parameters as needed.

```python
# Example Line Chart

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating the line chart
plt.plot(x, y)

# Adding labels to the x and y axes as well as the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example')

# Displaying the chart
plt.show()
```

In this example:

- We import the Matplotlib library as plt.
- We define two lists, x and y, to represent the x and y coordinates of the data points.
- We use the plt.plot() function to create the line chart. This function takes two arguments: the x values and the y values.
- We add labels to the x-axis, y-axis, and title of the chart using the plt.xlabel(), plt.ylabel(), and plt.title() functions respectively.
- Finally, we use the plt.show() function to display the chart.

```python
# Example Multiple Line Chart

import matplotlib.pyplot as plt

# Data for the first line
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]

# Data for the second line
x2 = [1, 2, 3, 4, 5]
y2 = [1, 3, 5, 7, 9]

# Creating the multi-line chart
plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')

# Adding labels to the x and y axes as well as the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multi-Line Chart Example')

# Adding a legend
plt.legend()

# Displaying the chart
plt.show()
```

In this example:

- We import the Matplotlib library as plt.
- We define two sets of data points: (x1, y1) for the first line and (x2, y2) for the second line.
- We use the plt.plot() function twice to create two lines on the same chart. Each plt.plot() call takes the x values, y values, and an optional label parameter to specify the label for the line.
- We add labels to the x-axis, y-axis, and title of the chart using the plt.xlabel(), plt.ylabel(), and plt.title() functions respectively.
- We add a legend to the chart using the plt.legend() function to differentiate between the two lines.
- Finally, we use the plt.show() function to display the chart.

## Composition

Composition refers to the arrangement or combination of various elements or parts that form a whole. In the context of data analytics or data visualization, composition is often used to describe how a whole is made up of or divided into smaller parts.

```python
# Example Pie Chart

import matplotlib.pyplot as plt

# Data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [20, 30, 25, 25]

# Creating the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Adding a title
plt.title('Pie Chart Example')

# Displaying the chart
plt.show()
```

A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. The arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents.

```python
# Example Stacked Bar Chart

import matplotlib.pyplot as plt

# Data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values1 = [20, 35, 30, 25]  # Values for the first category
values2 = [15, 25, 20, 30]  # Values for the second category

# Creating the stacked bar chart
plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')

# Adding labels to the x and y axes as well as the title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')

# Adding a legend
plt.legend()

# Displaying the chart
plt.show()
```

In this example:

- We import the Matplotlib library as plt.
- We define two sets of data, values1 and values2, representing the values for two different groups or categories.
- We use the plt.bar() function twice to create the stacked bar chart. The bottom parameter is used to specify the starting point for the second set of bars, making them stacked on top of the first set.
- We add labels to the x-axis, y-axis, and title of the chart using the plt.xlabel(), plt.ylabel(), and plt.title() functions respectively.
- We add a legend to differentiate between the two groups using the plt.legend() function.
- Finally, we use the plt.show() function to display the chart.

```python
# Example Stacked 100% Bar Chart

import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values1 = [20, 35, 30, 25]  # Values for the first category
values2 = [15, 25, 20, 30]  # Values for the second category

# Calculate the total sum for each category
total = np.array(values1) + np.array(values2)

# Calculate the percentage for each group
percentage1 = np.array(values1) / total * 100
percentage2 = np.array(values2) / total * 100

# Create the stacked 100% bar chart
plt.bar(categories, percentage1, label='Group 1')
plt.bar(categories, percentage2, bottom=percentage1, label='Group 2')

# Adding labels to the x and y axes as well as the title
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('Stacked 100% Bar Chart Example')

# Adding a legend
plt.legend()

# Displaying the chart
plt.show()
```

In this example:

- We import the Matplotlib library as plt and the NumPy library as np.
- We define two sets of data, values1 and values2, representing the values for two different groups or categories.
- We calculate the total sum for each category using NumPy.
- We calculate the percentage for each group by dividing each group's values by the total and multiplying by 100.
- We use the plt.bar() function twice to create the stacked 100% bar chart, using the calculated percentages as the bar heights.
- We add labels to the x-axis, y-axis, and title of the chart using the plt.xlabel(), plt.ylabel(), and plt.title() functions respectively.
- We add a legend to differentiate between the two groups using the plt.legend() function.
- Finally, we use the plt.show() function to display the chart.

```python
# Example Map

import folium

# Create a map centered at a specific location
m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)

# Add a marker to the map
folium.Marker(location=[51.5074, -0.1278], popup='London').add_to(m)

# Save the map to an HTML file
m.save('map.html')
```

In this example:

- We import the folium library, which is a Python wrapper for Leaflet.js, a JavaScript library for interactive maps.
- We create a map centered at a specific location using the folium.Map() function. The location parameter specifies the latitude and longitude of the center of the map, and the zoom_start parameter sets the initial zoom level.
- We add a marker to the map using the folium.Marker() function. The location parameter specifies the latitude and longitude of the marker, and the popup parameter sets the text that appears when the marker is clicked.
- We save the map to an HTML file using the save() method.
    
You can open the generated HTML file (map.html) in a web browser to view the map with the marker. You can customize the map further by adding additional markers, shapes, or layers, and by adjusting the map parameters as needed.
